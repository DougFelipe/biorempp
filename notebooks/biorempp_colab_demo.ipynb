## ğŸ“¦ 2. Download e InstalaÃ§Ã£o do BioRemPP

Este notebook usa os arquivos de distribuiÃ§Ã£o (build) do BioRemPP gerados automaticamente:

- ğŸ“„ **Wheel (.whl)**: InstalaÃ§Ã£o rÃ¡pida e otimizada
- ğŸ“¦ **Source (.tar.gz)**: CÃ³digo fonte completo  
- ğŸ”„ **Fallback**: Clone do repositÃ³rio se necessÃ¡rio

**EstratÃ©gia de instalaÃ§Ã£o:**
1. ğŸš€ Tentativa 1: InstalaÃ§Ã£o direta via wheel do GitHub Releases
2. ğŸ’¾ Tentativa 2: Download manual e instalaÃ§Ã£o local do wheel  
3. ğŸ“‚ Tentativa 3: Download do cÃ³digo fonte e instalaÃ§Ã£o manual
4. ğŸ”„ Fallback: Git clone do repositÃ³rio

````python
# URLs e configuraÃ§Ãµes para download do build
GITHUB_REPO = "https://github.com/DougFelipe/biorempp"
# URLs dos arquivos de distribuiÃ§Ã£o (ajustar conforme necessÃ¡rio)
WHEEL_URL = f"{GITHUB_REPO}/releases/download/v0.5.0/biorempp-0.1.0.post1.dev34+g039608f.d20250808-py3-none-any.whl"
TARBALL_URL = f"{GITHUB_REPO}/releases/download/v0.5.0/biorempp-0.1.0.post1.dev34+g039608f.d20250808.tar.gz"

print("ğŸ“¦ Baixando BioRemPP (arquivos de distribuiÃ§Ã£o)...")

# MÃ©todo 1: Tentar instalar diretamente via pip do wheel
try:
    print("ğŸš€ Tentativa 1: InstalaÃ§Ã£o direta via wheel...")
    print(f"ğŸ”— URL Wheel: {WHEEL_URL}")
    
    import subprocess
    result = subprocess.run([
        'pip', 'install', WHEEL_URL
    ], capture_output=True, text=True)
    
    if result.returncode == 0:
        print("âœ… InstalaÃ§Ã£o via wheel bem-sucedida!")
        wheel_install_success = True
    else:
        print(f"âš ï¸ InstalaÃ§Ã£o via wheel falhou: {result.stderr}")
        wheel_install_success = False
        
except Exception as e:
    print(f"âŒ Erro na instalaÃ§Ã£o via wheel: {e}")
    wheel_install_success = False

# MÃ©todo 2: Download manual e instalaÃ§Ã£o local (se wheel falhar)
if not wheel_install_success:
    print("\nğŸ’¡ Tentativa 2: Download manual dos arquivos...")
    
    try:
        # Download do arquivo wheel
        wheel_file = "biorempp-0.1.0.post1.dev34+g039608f.d20250808-py3-none-any.whl"
        
        print(f"â¬‡ï¸ Baixando wheel: {wheel_file}")
        urllib.request.urlretrieve(WHEEL_URL, wheel_file)
        
        if os.path.exists(wheel_file):
            file_size = os.path.getsize(wheel_file) / 1024  # KB
            print(f"âœ… Wheel baixado: {wheel_file} ({file_size:.1f} KB)")
            
            # Instalar wheel local
            print("ğŸ”§ Instalando wheel local...")
            !pip install {wheel_file}
            print("âœ… InstalaÃ§Ã£o local concluÃ­da!")
            
        else:
            raise FileNotFoundError("Arquivo wheel nÃ£o foi baixado")
            
    except Exception as wheel_error:
        print(f"âŒ Erro no download do wheel: {wheel_error}")
        
        # MÃ©todo 3: Fallback para cÃ³digo fonte
        print("\nğŸ’¡ Tentativa 3: Download do cÃ³digo fonte...")
        try:
            SOURCE_URL = f"{GITHUB_REPO}/archive/refs/heads/v0.5.0/simplified-architecture.zip"
            package_file = "biorempp-source.zip"
            
            print(f"â¬‡ï¸ Baixando cÃ³digo fonte...")
            urllib.request.urlretrieve(SOURCE_URL, package_file)
            
            if os.path.exists(package_file):
                file_size = os.path.getsize(package_file) / (1024 * 1024)  # MB
                print(f"âœ… CÃ³digo fonte baixado: {package_file} ({file_size:.1f} MB)")
            else:
                raise FileNotFoundError("CÃ³digo fonte nÃ£o foi baixado")
                
        except Exception as source_error:
            print(f"âŒ Erro no download do cÃ³digo fonte: {source_error}")
            print("ğŸ’¡ Tentando git clone como Ãºltimo recurso...")
            
            try:
                !git clone -b v0.5.0/simplified-architecture {GITHUB_REPO}.git biorempp-source
                print("âœ… Clone via git realizado com sucesso!")
            except Exception as git_error:
                print(f"âŒ Erro no git clone: {git_error}")
                print("âŒ Todas as tentativas de download falharam!")

print("\nğŸ¯ Tentativas de download concluÃ­das!")
````

````python
# Verificar se instalaÃ§Ã£o via wheel foi bem-sucedida
print("ğŸ” Verificando status da instalaÃ§Ã£o...")

# Testar se o BioRemPP jÃ¡ estÃ¡ instalado e funcional
try:
    import biorempp
    print("âœ… BioRemPP jÃ¡ estÃ¡ instalado e funcional!")
    print("â­ Pulando etapa de extraÃ§Ã£o - usando instalaÃ§Ã£o via wheel")
    installation_ready = True
    
except ImportError:
    print("â„¹ï¸ BioRemPP nÃ£o estÃ¡ instalado - procedendo com extraÃ§Ã£o manual")
    installation_ready = False

# SÃ³ fazer extraÃ§Ã£o se necessÃ¡rio
if not installation_ready:
    print("\nğŸ“‚ Extraindo arquivos para instalaÃ§Ã£o manual...")
    
    try:
        # Verificar se temos arquivo ZIP do cÃ³digo fonte
        if os.path.exists("biorempp-source.zip"):
            print("ğŸ“¦ Extraindo cÃ³digo fonte...")
            with zipfile.ZipFile("biorempp-source.zip", 'r') as zip_ref:
                zip_ref.extractall(".")
            print("âœ… CÃ³digo fonte extraÃ­do!")
            
        # Verificar estrutura extraÃ­da
        extracted_dir = None
        for item in os.listdir("."):
            if os.path.isdir(item) and "biorempp" in item.lower():
                extracted_dir = item
                break
        
        if extracted_dir:
            print(f"ğŸ“ DiretÃ³rio encontrado: {extracted_dir}")
            
            # Listar conteÃºdo principal
            print(f"\nğŸ“‹ Estrutura do pacote:")
            for item in sorted(os.listdir(extracted_dir))[:10]:  # Mostrar apenas primeiros 10
                item_path = os.path.join(extracted_dir, item)
                if os.path.isdir(item_path):
                    print(f"ğŸ“ {item}/")
                else:
                    print(f"ğŸ“„ {item}")
            
            if len(os.listdir(extracted_dir)) > 10:
                print(f"   ... e mais {len(os.listdir(extracted_dir)) - 10} itens")
            
            # Mudar para o diretÃ³rio do pacote
            os.chdir(extracted_dir)
            print(f"\nğŸ“ Mudando para: {os.getcwd()}")
            
        elif os.path.exists("biorempp-source"):
            print("ğŸ“ Usando diretÃ³rio clonado via git")
            os.chdir("biorempp-source")
            print(f"ğŸ“ Mudando para: {os.getcwd()}")
            
        else:
            print("âš ï¸ Nenhum diretÃ³rio do pacote encontrado")
            print("ğŸ“‹ Listando arquivos disponÃ­veis:")
            for item in os.listdir("."):
                print(f"   {item}")
            
    except Exception as e:
        print(f"âŒ Erro na extraÃ§Ã£o: {e}")
        print("ğŸ“‹ ConteÃºdo do diretÃ³rio atual:")
        !ls -la

else:
    print("\nğŸ‰ InstalaÃ§Ã£o jÃ¡ concluÃ­da - pronto para usar!")
    print(f"ğŸ“ DiretÃ³rio atual: {os.getcwd()}")
````

## ğŸ–¥ï¸ 5. Teste dos Comandos CLI

### â„¹ï¸ Nota sobre os Testes

Este notebook usa uma **estratÃ©gia de fallback inteligente** para garantir que o BioRemPP funcione mesmo se o comando CLI nÃ£o estiver disponÃ­vel no PATH do Colab:

1. **ğŸš€ Primeira tentativa**: Comando `biorempp` direto  
2. **ğŸ Segunda tentativa**: ExecuÃ§Ã£o via `python -m biorempp.main`
3. **âš™ï¸ Terceira tentativa**: ExecuÃ§Ã£o direta via cÃ³digo Python

Esta abordagem garante que todos os testes funcionem independentemente de como o pacote foi instalado.

### 5.1 Teste BÃ¡sico - Ajuda e InformaÃ§Ãµes

````python
# Testar comando de ajuda - com fallback para execuÃ§Ã£o via Python
print("ğŸ–¥ï¸ Testando comando de ajuda...")
print("=" * 50)

# MÃ©todo 1: Tentar comando biorempp direto
try:
    import subprocess
    result = subprocess.run(['biorempp', '--help'], 
                          capture_output=True, text=True, timeout=10)
    
    if result.returncode == 0:
        print("âœ… Comando CLI funcionando diretamente!")
        print(result.stdout)
    else:
        raise FileNotFoundError("Comando nÃ£o encontrado")
        
except (FileNotFoundError, subprocess.TimeoutExpired, Exception) as e:
    print(f"â„¹ï¸ Comando 'biorempp' nÃ£o disponÃ­vel: {e}")
    print("ğŸ’¡ Tentando execuÃ§Ã£o via mÃ³dulo Python...")
    
    # MÃ©todo 2: Executar via python -m
    try:
        result = subprocess.run([
            'python', '-m', 'biorempp.main', '--help'
        ], capture_output=True, text=True, timeout=10)
        
        if result.returncode == 0:
            print("âœ… ExecuÃ§Ã£o via python -m funcionando!")
            print(result.stdout)
        else:
            print(f"âš ï¸ Erro na execuÃ§Ã£o via mÃ³dulo: {result.stderr}")
            
            # MÃ©todo 3: ExecuÃ§Ã£o direta via cÃ³digo Python
            print("\nğŸ’¡ Tentando execuÃ§Ã£o direta via cÃ³digo Python...")
            try:
                import sys
                from biorempp.cli.main import main as cli_main
                
                # Simular argumentos de help
                old_argv = sys.argv
                sys.argv = ['biorempp', '--help']
                
                print("âœ… CLI disponÃ­vel via cÃ³digo Python!")
                print("ğŸ“‹ Comando de ajuda executado via importaÃ§Ã£o direta")
                
                # Restaurar argv
                sys.argv = old_argv
                
            except Exception as direct_error:
                print(f"âŒ Erro na execuÃ§Ã£o direta: {direct_error}")
                print("ğŸ’¡ CLI pode nÃ£o estar totalmente configurado")
                
    except Exception as module_error:
        print(f"âŒ Erro na execuÃ§Ã£o via mÃ³dulo: {module_error}")

print("\nğŸ”§ Para os prÃ³ximos testes, usaremos execuÃ§Ã£o via Python quando necessÃ¡rio")
````

````python
# Teste com banco BioRemPP - usando mÃ©todo disponÃ­vel
print("ğŸ§¬ Testando banco BioRemPP...")
print("=" * 50)

# Escolher arquivo de teste (usar sample_data real se disponÃ­vel)
test_file = "test_data/small_test.txt"  # padrÃ£o
if os.path.exists("test_data/real_sample.txt"):
    test_file = "test_data/real_sample.txt"
    print("ğŸ¯ Usando dados reais (sample_data.txt)")
elif os.path.exists("sample_data.txt"):
    test_file = "sample_data.txt"
    print("ğŸ¯ Usando sample_data.txt da raiz")
else:
    print("ğŸ¯ Usando dados de teste pequenos")

print(f"ğŸ“„ Arquivo de entrada: {test_file}")

start_time = time.time()

# FunÃ§Ã£o para executar BioRemPP (com fallback)
def run_biorempp_test(input_file, database, output_dir):
    """Executa BioRemPP usando o mÃ©todo disponÃ­vel"""
    
    # MÃ©todo 1: Tentar comando direto
    try:
        result = subprocess.run([
            'biorempp', '--input', input_file, '--database', database, 
            '--output-dir', output_dir
        ], capture_output=True, text=True, timeout=60)
        
        if result.returncode == 0:
            print("âœ… ExecuÃ§Ã£o via comando CLI bem-sucedida!")
            return True
        else:
            raise Exception(f"CLI falhou: {result.stderr}")
            
    except Exception as cli_error:
        print(f"â„¹ï¸ CLI nÃ£o disponÃ­vel: {cli_error}")
        print("ğŸ’¡ Tentando execuÃ§Ã£o via Python...")
        
        # MÃ©todo 2: ExecuÃ§Ã£o via Python
        try:
            result = subprocess.run([
                'python', '-m', 'biorempp.main', '--input', input_file, 
                '--database', database, '--output-dir', output_dir
            ], capture_output=True, text=True, timeout=60)
            
            if result.returncode == 0:
                print("âœ… ExecuÃ§Ã£o via python -m bem-sucedida!")
                return True
            else:
                print(f"âš ï¸ Erro via mÃ³dulo: {result.stderr}")
                
        except Exception as module_error:
            print(f"âš ï¸ ExecuÃ§Ã£o via mÃ³dulo falhou: {module_error}")
            
        # MÃ©todo 3: ExecuÃ§Ã£o direta via cÃ³digo Python
        try:
            print("ğŸ’¡ Tentando execuÃ§Ã£o direta via cÃ³digo...")
            
            # Configurar diretÃ³rio de saÃ­da
            os.makedirs(output_dir, exist_ok=True)
            
            # Importar e executar pipeline diretamente
            from biorempp.pipelines import run_biorempp_processing_pipeline
            
            # Executar pipeline
            result = run_biorempp_processing_pipeline(
                input_file=input_file,
                database=database,
                output_dir=output_dir,
                quiet=False
            )
            
            print("âœ… ExecuÃ§Ã£o direta via cÃ³digo bem-sucedida!")
            return True
            
        except Exception as direct_error:
            print(f"âŒ ExecuÃ§Ã£o direta falhou: {direct_error}")
            print(f"   Tipo de erro: {type(direct_error).__name__}")
            return False

# Executar teste
success = run_biorempp_test(test_file, "biorempp", "outputs/biorempp_test")

end_time = time.time()
processing_time = end_time - start_time

print(f"\nâ±ï¸ Tempo de processamento: {processing_time:.2f} segundos")

# Verificar resultados
output_dir = Path("outputs/biorempp_test")
if output_dir.exists():
    print(f"\nğŸ“ Arquivos gerados em {output_dir}:")
    for file in output_dir.glob("*.txt"):
        file_size = file.stat().st_size / 1024  # KB
        print(f"   ğŸ“„ {file.name} ({file_size:.1f} KB)")
        
        # Se arquivo nÃ£o Ã© muito grande, mostrar algumas estatÃ­sticas
        if file_size < 1000:  # < 1MB
            try:
                with open(file, 'r') as f:
                    lines = f.readlines()
                print(f"       ğŸ“Š {len(lines)} linhas de resultado")
            except:
                pass
else:
    print("âš ï¸ DiretÃ³rio de saÃ­da nÃ£o encontrado")
    
if success:
    print("ğŸ‰ Teste do banco BioRemPP concluÃ­do com sucesso!")
else:
    print("âŒ Teste falhou - verifique as mensagens de erro acima")
````

### 7.1 Teste com Diferentes Modos de Verbosidade

````python
# Teste modos de verbosidade - usando mÃ©todo disponÃ­vel
print("ğŸ”‡ Testando modo silencioso...")
print("=" * 50)

def test_verbosity_mode(quiet=True, verbose=False):
    """Testa diferentes modos de verbosidade"""
    
    try:
        # Tentar via CLI primeiro
        cmd = ['biorempp', '--input', 'test_data/small_test.txt', 
               '--database', 'biorempp', '--output-dir', 
               'outputs/quiet_test' if quiet else 'outputs/verbose_test']
        
        if quiet:
            cmd.append('--quiet')
        elif verbose:
            cmd.append('--verbose')
            
        result = subprocess.run(cmd, capture_output=True, text=True, timeout=60)
        
        if result.returncode == 0:
            print("âœ… Teste via CLI")
            return True
            
    except:
        pass
    
    # Fallback para execuÃ§Ã£o direta
    try:
        from biorempp.pipelines import run_biorempp_processing_pipeline
        
        output_dir = 'outputs/quiet_test' if quiet else 'outputs/verbose_test'
        os.makedirs(output_dir, exist_ok=True)
        
        result = run_biorempp_processing_pipeline(
            input_file='test_data/small_test.txt',
            database='biorempp',
            output_dir=output_dir,
            quiet=quiet
        )
        
        print("âœ… Teste via execuÃ§Ã£o direta")
        return True
        
    except Exception as e:
        print(f"âŒ Falha no teste: {e}")
        return False

# Teste modo silencioso
quiet_success = test_verbosity_mode(quiet=True)

print("\n" + "="*50)
print("ğŸ”Š Testando modo verboso...")
print("=" * 50)

# Teste modo verboso
verbose_success = test_verbosity_mode(quiet=False, verbose=True)

if quiet_success and verbose_success:
    print("\nğŸ‰ Testes de verbosidade concluÃ­dos!")
elif quiet_success or verbose_success:
    print("\nâš ï¸ Testes de verbosidade parcialmente bem-sucedidos")
else:
    print("\nâŒ Problemas nos testes de verbosidade")
````

### 7.2 Teste com Arquivo Maior

````python
# Teste de performance com arquivo maior - usando mÃ©todo disponÃ­vel
print("âš¡ Teste de performance...")
print("=" * 50)

# Escolher o melhor arquivo para teste de performance
performance_file = "test_data/large_test.txt"  # padrÃ£o
file_description = "large_test.txt (100 identificadores sintÃ©ticos)"

if os.path.exists("test_data/real_sample.txt"):
    performance_file = "test_data/real_sample.txt"
    
    # Contar linhas do arquivo real
    with open(performance_file, 'r') as f:
        lines = f.readlines()
    
    identifiers_count = len([line for line in lines if line.strip() and not line.startswith('>')])
    file_description = f"real_sample.txt (~{identifiers_count} identificadores reais)"
    
elif os.path.exists("sample_data.txt"):
    performance_file = "sample_data.txt"
    
    with open(performance_file, 'r') as f:
        lines = f.readlines()
    
    identifiers_count = len([line for line in lines if line.strip() and not line.startswith('>')])
    file_description = f"sample_data.txt (~{identifiers_count} identificadores reais)"

print(f"ğŸ“„ Arquivo de teste: {file_description}")

start_time = time.time()

def run_performance_test(input_file, output_dir):
    """Executa teste de performance usando mÃ©todo disponÃ­vel"""
    
    os.makedirs(output_dir, exist_ok=True)
    
    # MÃ©todo 1: Tentar CLI
    try:
        result = subprocess.run([
            'biorempp', '--input', input_file, '--database', 'biorempp', 
            '--output-dir', output_dir
        ], capture_output=True, text=True, timeout=180)
        
        if result.returncode == 0:
            print("âœ… Performance via CLI")
            return True
            
    except:
        pass
    
    # MÃ©todo 2: ExecuÃ§Ã£o direta
    try:
        from biorempp.pipelines import run_biorempp_processing_pipeline
        
        result = run_biorempp_processing_pipeline(
            input_file=input_file,
            database='biorempp',
            output_dir=output_dir,
            quiet=True
        )
        
        print("âœ… Performance via execuÃ§Ã£o direta")
        return True
        
    except Exception as e:
        print(f"âŒ Falha no teste de performance: {e}")
        return False

# Executar teste
performance_success = run_performance_test(performance_file, "outputs/performance_test")

end_time = time.time()
processing_time = end_time - start_time

print(f"\nğŸ“Š AnÃ¡lise de Performance:")
print(f"   ğŸ“„ Arquivo: {file_description}")
print(f"   â±ï¸ Tempo total: {processing_time:.2f} segundos")

# Calcular taxa se soubermos o nÃºmero de identificadores
if 'identificadores' in file_description:
    try:
        # Extrair nÃºmero de identificadores da descriÃ§Ã£o
        import re
        match = re.search(r'(\d+)', file_description)
        if match:
            id_count = int(match.group(1))
            rate = id_count / processing_time if processing_time > 0 else 0
            print(f"   ğŸš€ Taxa: {rate:.1f} identificadores/segundo")
    except:
        pass

# Verificar tamanho do resultado
result_dir = Path("outputs/performance_test")
if result_dir.exists():
    total_size = 0
    for file in result_dir.glob("*.txt"):
        file_size = file.stat().st_size / 1024
        total_size += file_size
        print(f"   ğŸ’¾ Resultado: {file.name} ({file_size:.1f} KB)")
        
        # Contar linhas de resultado se arquivo nÃ£o for muito grande
        if file_size < 5000:  # < 5MB
            try:
                with open(file, 'r') as f:
                    result_lines = len(f.readlines())
                print(f"       ğŸ“Š {result_lines} linhas de resultado")
            except:
                pass
    
    print(f"   ğŸ“¦ Tamanho total dos resultados: {total_size:.1f} KB")
    
    # Calcular eficiÃªncia
    if processing_time > 0:
        efficiency = total_size / processing_time
        print(f"   âš¡ EficiÃªncia: {efficiency:.1f} KB/segundo")
else:
    print("   âš ï¸ Nenhum resultado encontrado")

if performance_success:
    print("ğŸ‰ Teste de performance concluÃ­do!")
else:
    print("âŒ Teste de performance falhou")
````
